```python
import pandas as pd
import numpy as np
import statistics
import math
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sbi_data.csv', usecols = ['Date', 'Close'])
```

```python
for i in range(66, len(df.Close)):
    df.Close[i] = 0.00
#     print(df.Close[i])
```

```python
df
```

```python
U = []
for i in range(65):
    U.append( math.log(df.Close[i+1]/df.Close[i]) )
```

```python
variance = statistics.variance(U)
sigma = math.sqrt(variance)
print(variance)
```

```python
mean = statistics.mean(U)
print(mean)
```

```python
mu = variance/2 + mean
print(mu)
```

```python
S = []
```

```python
# t1 -> t(i)
# t2 -> t(i+1)
def predict(t1, t2):
    S[t2] = S[t1] * math.exp( (mu-variance/2)*(t2-t1) + sigma * (t2-t1) * np.random.normal(0.00, 1.00))
    return S[t2]
```

```python
# 7th of October is Day Number 5 (Day 70 in CSV)
# 14th of October is Day Number 10 (Day 75 in CSV)
# 21st of October is Day Number 15 (Day 80 in CSV)
# *** Saturday - Sundays are off***
```

```python
data_1 = []
data_2 = []
data_3 = []
delta = 1

data = []
for i in range(99):
    data.append(0.00)

N = 10000
for ctr in range(N):
    S.clear()
    S.append(-1)
    S.append(df.Close[65])
    
    for i in range(delta+1, 16):
#         print(i-delta, i)
        S.append(-1)
        S[i] = predict(i-delta, i)
    
    for j in range(len(S)-1):
        data[j] += S[j+1]
    
#     print(S)
    data_1.append(S[5])
    data_2.append(S[10])
    data_3.append(S[15])    

while data[-1] == 0.00:
    data.pop()
    
for j in range(len(data)):
    data[j] /= N
```

```python
df = pd.read_csv('sbi_data.csv', usecols = ['Date', 'Close'])
```

```python
# print(data_1)
print(len(data_1))
print(statistics.mean(data_1), statistics.mean(data_2), statistics.mean(data_3))
print(df.Close[69], df.Close[74], df.Close[79])
```

```python
key = []
for i in range(65):
    key.append(df.Close[i])

for i in data:
    key.append(i)
    
predicted = pd.Series(key)

%matplotlib inline
predicted.plot(kind='hist')
```

```python
df['Predicted_Close'] = predicted
```

```python
df['Error %'] = ((abs(df['Predicted_Close'] - df['Close']))/df['Close'])*100
```
